```python
local_file = 'GPS_GUI_Data_2023-04-26_16:14:44_R.csv'
data1 = pd.read_csv(local_file)
data2 = pd.read_csv('GPS_GUI_Data_2023-04-26_16:42:12_R.csv')
data3 = pd.read_csv('GPS_GUI_Data_2023-04-26_18:46:05_R.csv')

df = pd.concat(
    map(pd.read_csv, ['GPS_GUI_Data_2023-04-26_16:14:44_R.csv', 'GPS_GUI_Data_2023-04-26_16:42:12_R.csv', 'GPS_GUI_Data_2023-04-26_18:46:05_R.csv']), ignore_index=True)
display(df)
```

```python
m = leafmap.Map(tiles='stamentoner')
m.add_heatmap(
    df,
    latitude="Latitude",
    longitude='Longitude',
    value="Radiation (cps)",
    name="Heat map",
    radius=10,
)
m
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data file
df = data1

# Calculate the average count
avg_count = df['Radiation (cps)'].mean()

# Plot the average count as a horizontal line
plt.axhline(y=avg_count, color='r', linestyle='-', label='Average Count')

# Plot the count vs time graph
plt.plot(df['Epoch time'], df['Radiation (cps)'], label='Radiation (cps) vs Time')

# Add a title and axis labels
plt.title('Radiation (cps) vs Time')
plt.xlabel('Time')
plt.ylabel('Radiation (cps)')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```
